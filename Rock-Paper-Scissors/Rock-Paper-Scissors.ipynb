```python
import random
user_continue = "y"
dic = {"rock": 0, "paper": 1, "scissors": 2}
user_input = str()
computer_point = 0
user_point = 0

while user_continue == "y":
    #restrict the input
    while user_input != "r" and user_input != "p" and user_input !="s":
        user_input = str(input("Please type r for rock, p for paper, or s for scissors: "))
        #clean input
        if user_input == "r":
            user_choice = "rock"
        elif user_input == "p":
            user_choice = "paper"
        elif user_input == "s":
            user_choice = "scissors"
        else:
            print("Invalid input. It has to be r for rock, p for paper, or s for scissors\n")
    
    #computer makes a random choice
    computer_choice = random.choice(["rock","paper","scissors"])
    print("Computer's choice is: " + computer_choice)
    
    #cases of rock, paper, scissors
    if dic[user_choice] == dic[computer_choice] - 1 or dic[user_choice] == dic[computer_choice] + 2:
        computer_point += 1
        print("You lose.")
    elif dic[user_choice] == dic[computer_choice]:
        print("Tie.")
    else:
        user_point += 1
        print("You win.")
    user_input = str()
    user_continue = str()
    print("You: " + str(user_point) + " Computer: " + str(computer_point))
    
    while user_continue != "y" and user_continue != "n":
        user_continue = str(input("Do you want to continue? Please type y or n: "))
        if user_continue != "y" and user_continue != "n":
            print("Invalid input. Please type y for yes and n for no")
        
    print()
```
