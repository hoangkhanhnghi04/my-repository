```python
class BlackJack():
    def __init__(self,card_list):
        self.card_list = card_list
        
    def Score_Count(self):
        sum_cards = 0
        #transform all character cards to 10
        character_cards = ["Jack","Queen","King"]
        for i in range(0,len(self.card_list)):
            if self.card_list[i] in character_cards:
                sum_cards += 10
            elif self.card_list[i] not in character_cards and self.card_list[i] != "Ace":
                sum_cards += self.card_list[i]
            else:
                sum_cards += 1
            
        if sum_cards >= 22:
            return sum_cards
        elif sum_cards == 12 and "Ace" in self.card_list:
            sum_cards += 9
        elif sum_cards < 12 and sum_cards > 2 and "Ace" in self.card_list:
            sum_cards += 10
        elif sum_cards == 2:
            sum_cards = 21
        return sum_cards
    
    def Check_Pair_Of_Aces(self):
        #check xi hoang
        if len(self.card_list) == 2 and self.card_list[0] == "Ace" and self.card_list[1] == "Ace":
            return True
        return False

    def Check_Natural(self):
        #check xi zach
        ten_list = [10,"Jack","Queen","King"]
        if len(self.card_list) == 2 and "Ace" in self.card_list and (self.card_list[0] in ten_list or self.card_list[1] in ten_list):
            return True
        return False
    
    def Check_Five_Cards(self):
        #check ngu linh
        if len(self.card_list) == 5 and self.Score_Count() <= 21:
            return True
        return False

```

```python
def compare(computer_list, user_list):
    computer_obj = BlackJack(computer_list)
    user_obj = BlackJack(user_list)
    computer_point = computer_obj.Score_Count()
    user_point = user_obj.Score_Count()
    print("Computer's point: " + str(computer_point) + " . Your point: " + str(user_point))
    
    #xi hoang
    if (computer_obj.Check_Pair_Of_Aces() and not user_obj.Check_Pair_Of_Aces()):
        return("You lose")
    elif not computer_obj.Check_Pair_Of_Aces() and user_obj.Check_Pair_Of_Aces():
        return("You win")
    
    #ngu linh
    elif computer_obj.Check_Five_Cards() and not user_obj.Check_Five_Cards():
        return("You lose")
    elif computer_obj.Check_Five_Cards() and user_obj.Check_Five_Cards():
        if computer_point < user_point:
            return("You lose")
        else:
            return("You win")
    elif not computer_obj.Check_Five_Cards() and user_obj.Check_Five_Cards():
        return("You win")
    
    #xi zach
    elif (computer_obj.Check_Natural() and not user_obj.Check_Natural()):
        return("You lose")
    elif (computer_obj.Check_Natural() and user_obj.Check_Natural()):
        return("You win")
    
    #khong phai case dac biet
    else:
        if user_point < 15:
            return("Less than 15. You lose.")
        elif user_point >= 22:
            if computer_point >= 22:
                return("Tie.")
            else:
                return("You lose")
        else:
            if computer_point >= 22:
                return("You win")
            elif computer_point > user_point:
                return("You lose")
            elif computer_point == user_point:
                return("Tie")
            else:
                return("You win")
```

```python
import random
card_deck = ["Ace",2,3,4,5,6,7,8,9,10,"Jack","Queen","King"]*4
computer_cards = random.choices(card_deck,k=2)
for i in computer_cards:
    card_deck.remove(i)

while BlackJack(computer_cards).Score_Count() < 16 and len(computer_cards)<5:
    computer_next_card = random.choice(card_deck)
    computer_cards.append(computer_next_card)
    card_deck.remove(computer_next_card)

user_choice = input("Do you want to play Blackjack? Please type \"y\" to start: ")

if user_choice == "y":
    user_cards = random.choices(card_deck,k=2)
    for i in user_cards:
        card_deck.remove(i)
    print("Your first card is: " + str(user_cards[0]))
    print("Your second card is: " + str(user_cards[1]))
    
    user_withdraw = str()
    while user_withdraw != "y" and user_withdraw != "n":
        user_withdraw = input("Do you want to withdraw another card? Please type \"y\" or \"n\": ")
        while user_withdraw == "y":
            next_card = random.choice(card_deck)
            user_cards.append(next_card)
            card_deck.remove(next_card)
            print("Your new card is " + str(next_card))
            user_withdraw = str()

    print("\nComputer's cards are: " + str(computer_cards)[1:-1])
    print("Your cards are: " + str(user_cards)[1:-1])

    print(compare(computer_cards,user_cards))
```
